<h3>SemEval 2018 Affect</h3>


Semeval-2018 Task 1: Affect in Tweets. Saif M. Mohammad, Felipe Bravo-Marquez, Mohammad Salameh, and Svetlana Kiritchenko. In Proceedings of International Workshop on Semantic Evaluation (SemEval-2018), New Orleans, LA, USA, June 2018.

```python
path='data/EI-oc-En-train/'
semeval_anger=pd.read_csv(path+'EI-oc-En-anger-train.txt', sep='\t')
```

```python
semeval_anger['Intensity Class']=semeval_anger['Intensity Class'].str.split(':')
semeval_anger['Intensity Class']=semeval_anger['Intensity Class'].apply(lambda x: x[0])
semeval_anger['Intensity Class']=semeval_anger['Intensity Class'].astype(int)
semeval_train_anger=semeval_anger[semeval_anger['Intensity Class']>2]
```

```python
semeval_fear=pd.read_csv(path+'EI-oc-En-fear-train.txt', sep='\t')
```

```python
semeval_fear['Intensity Class']=semeval_fear['Intensity Class'].str.split(':')
semeval_fear['Intensity Class']=semeval_fear['Intensity Class'].apply(lambda x: x[0])
semeval_fear['Intensity Class']=semeval_fear['Intensity Class'].astype(int)
semeval_fear=semeval_fear[semeval_fear['Intensity Class']>1]
semeval_train_fear=semeval_fear
```

```python
semeval_joy=pd.read_csv(path+'EI-oc-En-joy-train.txt', sep='\t')
```

```python
semeval_joy['Intensity Class']=semeval_joy['Intensity Class'].str.split(':')
semeval_joy['Intensity Class']=semeval_joy['Intensity Class'].apply(lambda x: x[0])
semeval_joy['Intensity Class']=semeval_joy['Intensity Class'].astype(int)
semeval_joy=semeval_joy[semeval_joy['Intensity Class']>1]
```

```python
semeval_sadness=pd.read_csv(path+'EI-oc-En-sadness-train.txt', sep='\t')
```

```python
semeval_sadness['Intensity Class']=semeval_sadness['Intensity Class'].str.split(':')
semeval_sadness['Intensity Class']=semeval_sadness['Intensity Class'].apply(lambda x: x[0])
semeval_sadness['Intensity Class']=semeval_sadness['Intensity Class'].astype(int)
semeval_sadness=semeval_sadness[semeval_sadness['Intensity Class']>1]
```

```python
test_path='data/test-gold/EI-oc/'
semeval_test_anger=pd.read_csv(test_path+'2018-EI-oc-En-anger-test-gold.txt', sep='\t')
```

```python
semeval_anger=semeval_test_anger
semeval_anger['Intensity Class']=semeval_anger['Intensity Class'].str.split(':')
semeval_anger['Intensity Class']=semeval_anger['Intensity Class'].apply(lambda x: x[0])
semeval_anger['Intensity Class']=semeval_anger['Intensity Class'].astype(int)
semeval_1s=semeval_anger[semeval_anger['Intensity Class']==1].iloc[:100]
semeval_test_anger=semeval_anger[semeval_anger['Intensity Class']>2]
semeval_test_anger=semeval_test_anger.append(semeval_1s)
```

```python
len(semeval_test_anger)
```

```python
semeval_anger=semeval_train_anger.append(semeval_test_anger)
```

```python
len(semeval_anger)
```

```python
semeval_test_fear=pd.read_csv(test_path+'2018-EI-oc-En-fear-test-gold.txt', sep='\t')
```

```python
semeval_fear=semeval_test_fear
```

```python
semeval_fear['Intensity Class']=semeval_fear['Intensity Class'].str.split(':')
semeval_fear['Intensity Class']=semeval_fear['Intensity Class'].apply(lambda x: x[0])
semeval_fear['Intensity Class']=semeval_fear['Intensity Class'].astype(int)
semeval_test_fear=semeval_fear[semeval_fear['Intensity Class']>1]
```

```python
semeval_fear=semeval_train_fear.append(semeval_test_fear)
semeval_fear
```

```python
emotions=['anger', 'joy', 'fear', 'sadness']
dfs=[semeval_anger, semeval_joy, semeval_fear, semeval_sadness]
new_dfs=[]
for index, df in enumerate(dfs):
    sem_df=pd.DataFrame()
    sem_df[0]=emotions[index]
    sem_df[1]=df['Tweet']
    sem_df[0].replace(np.nan, emotions[index], inplace=True)  
    new_dfs.append(sem_df)
new_dfs
```

```python
semeval_emotions=pd.DataFrame()
for df in new_dfs:
    semeval_emotions=semeval_emotions.append(df)
semeval_emotions
```

```python
semeval_emotions.to_csv('data/semeval_emotions.csv', index=False)
```
